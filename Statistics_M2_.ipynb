1.What is hypothesis testing in statistics?

Hypothesis testing is a statistical method used to make inferences or decisions about a population based on sample data. It helps determine whether a claim (hypothesis) about a population parameter is likely to be true or false.

Key Steps in Hypothesis Testing:
1.State the Hypotheses

Null Hypothesis ( 𝐻 0 H 0​): The default assumption (e.g., no effect, no difference).

Alternative Hypothesis ( 𝐻 𝑎 H a​): The claim we want to test (e.g., there is an effect, a difference exists).

2.Choose the Significance Level ( 𝛼 α)

Common choices: 0.05 (5%), 0.01 (1%)

It represents the probability of rejecting 𝐻 0 H 0​when it is actually true (Type I error).

3.Select the Test and Compute the Test Statistic

Choose a test (e.g., t-test, chi-square test, ANOVA) based on data type and hypothesis.

Compute the test statistic, which measures how much the sample data deviates from 𝐻 0 H 0​.

4.Find the P-value or Critical Value

P-value: Probability of obtaining a result as extreme as the observed one under 𝐻 0 H 0​.

Compare with 𝛼 α: If p-value < 𝛼 α, reject 𝐻 0 H 0​; otherwise, fail to reject 𝐻 0 H 0​.

5.Draw a Conclusion

If 𝐻 0 H 0​is rejected, evidence supports 𝐻 𝑎 H a​.

If 𝐻 0 H 0​is not rejected, there is insufficient evidence to support 𝐻 𝑎 H a​.

2.What is the null hypothesis, and how does it differ from the alternative hypothesis?

The Null Hypothesis ( 𝐻 0 H 0​) and the Alternative Hypothesis ( 𝐻 𝑎 H a​or 𝐻 1 H 1​) are two competing statements in hypothesis testing.

Null Hypothesis ( 𝐻 0 H 0​)
The null hypothesis represents the default assumption that there is no effect, no difference, or no relationship in the population.

It is what we assume to be true unless we have enough evidence to reject it.

It is usually stated with "=" (equal sign), such as:

The alternative hypothesis is the claim we are testing. It suggests that there is an effect, a difference, or a relationship.

It contradicts 𝐻 0 H 0​and is supported if we find significant evidence against 𝐻 0 H 0​.

It often uses "≠", ">", or "<" signs:

3.What is the significance level in hypothesis testing, and why is it important?

Significance Level ( 𝛼 α) in Hypothesis Testing
The significance level ( 𝛼 α) is the probability of rejecting the null hypothesis ( 𝐻 0 H 0​) when it is actually true. In other words, it represents the risk of making a Type I error (false positive).

1.Common Choices for 𝛼 α:

0.05 (5%) → Most commonly used in research

0.01 (1%) → Used for stricter testing (e.g., medical studies)

0.10 (10%) → Used in exploratory research where some risk of error is acceptable

2.Why is the Significance Level Important?

Controls Type I Error: It limits the probability of mistakenly rejecting 𝐻 0 H 0​.

Determines Decision Rule: If the p-value (calculated from sample data) is less than 𝛼 α, we reject 𝐻 0 H 0​.

Affects Statistical Power: Lower 𝛼 α reduces the risk of Type I errors but increases the risk of Type II errors (failing to reject a false 𝐻 0 H 0​).

4.What does a P-value represent in hypothesis testing?

What is a P-Value in Hypothesis Testing?

The p-value is the probability of obtaining a test result at least as extreme as the observed one, assuming the null hypothesis ( 𝐻 0 H 0​) is true.

Interpretation of P-Value:

A small p-value (e.g., p < 0.05) suggests that the observed data is unlikely under 𝐻 0 H 0​, so we may reject 𝐻 0 H 0​.

A large p-value (e.g., p > 0.05) suggests that the data is consistent with 𝐻 0 H 0​, so we fail to reject it.

Key Points:

P-value is NOT the probability that 𝐻 0 H 0​is true.

It does NOT prove 𝐻 0 H 0​or 𝐻 𝑎 H a​, only provides evidence against 𝐻 0 H 0​.

Lower p-values indicate stronger evidence against 𝐻 0 H 0​.

Decision Rule Based on P-Value and Significance Level ( 𝛼 α)

If p-value < 𝛼 α: Reject 𝐻 0 H 0​(significant result).

If p-value ≥ 𝛼 α: Fail to reject 𝐻 0 H 0​(not enough evidence).

6.What are Type 1 and Type 2 errors in hypothesis testing?

Type I and Type II Errors in Hypothesis Testing

In hypothesis testing, errors occur when our conclusion about the null hypothesis ( 𝐻 0 H 0​) is incorrect. The two types of errors are:

1. Type I Error (False Positive)

Occurs when we reject 𝐻 0 H 0​when it is actually true.

It is like a false alarm—detecting an effect that doesn’t exist.

The probability of making a Type I error is the significance level ( 𝛼 α), often set at 0.05 (5%).

🔹 Example:

A drug company tests a new medicine to see if it lowers blood pressure.

𝐻 0 H 0​: The drug has no effect.

𝐻 𝑎 H a​: The drug lowers blood pressure.

Type I Error: We conclude the drug works when it actually doesn’t.

Consequence: Patients may take an ineffective drug, wasting money and risking side effects.

2. Type II Error (False Negative)

Occurs when we fail to reject 𝐻 0 H 0​when it is actually false.

It means missing a real effect—not detecting something that actually exists.

The probability of making a Type II error is 𝛽 β, which depends on sample size and effect size.

🔹 Example:

In the same drug study:

𝐻 0 H 0​: The drug has no effect.

𝐻 𝑎 H a​: The drug lowers blood pressure.

Type II Error: We conclude the drug doesn’t work when it actually does.

Consequence: A helpful drug is not approved, and patients miss out on treatment.

13.What is the relationship between Z-test and T-test in hypothesis testing?

Relationship Between Z-Test and T-Test in Hypothesis Testing

Both the Z-test and T-test are statistical tests used to compare sample data to a population parameter (typically the mean) in hypothesis testing. While they share many similarities, there are key differences that determine which test is appropriate to use. Here's a breakdown of the relationship and how they compare:

1. Similarities Between Z-Test and T-Test

Purpose: Both tests are used to test hypotheses about population means and help determine whether the observed data is significantly different from the expected population parameter.

Test Statistic: Both the Z-test and T-test calculate a test statistic (Z or T) that quantifies how far the sample mean is from the population mean in terms of standard errors.

Hypothesis Testing: Both tests follow the same process for hypothesis testing:

State the null hypothesis ( 𝐻 0 H 0​) and alternative hypothesis ( 𝐻 𝑎 H a​).

Choose a significance level ( 𝛼 α).

Calculate the test statistic.

Compare the test statistic to the critical value or use the p-value to make a decision.

Distribution: Both tests are based on probability distributions (Z-distribution for Z-test and T-distribution for T-test). The T-distribution is used for small samples, while the Z-distribution is used for large samples.

2.When to Use Z-Test vs. T-Test

Z-Test:

When sample size is large ( 𝑛 ≥ 30 n≥30): The Z-test is appropriate for large sample sizes because the Central Limit Theorem ensures that the sampling distribution of the sample mean will approach a normal distribution, even if the population itself is not normally distributed.

When population standard deviation is known ( 𝜎 σ): If the population standard deviation is available, the Z-test is the most efficient test to use because it uses the exact standard deviation of the population.

T-Test:

When sample size is small ( 𝑛 < 30 n<30): The T-test is more appropriate for small samples because the sample standard deviation is used as an estimate of the population standard deviation, and the T-distribution compensates for this uncertainty with heavier tails.

When population standard deviation is unknown: In most real-world situations, we don’t have access to the population standard deviation, so the T-test becomes the default choice when 𝜎 σ is unknown.

3.As Sample Size Increases, T-Test Approaches Z-Test

Convergence: As the sample size increases, the T-distribution approaches the normal distribution, meaning that for large samples, the T-test results become nearly identical to the Z-test results.

The heavier tails of the T-distribution adjust for the increased uncertainty in small samples, but as the sample size grows, this uncertainty reduces and the T-distribution becomes less "heavy," thus approaching the Z-distribution.

A confidence interval (CI) is a range of values that is used to estimate an unknown population parameter, such as a population mean or proportion, based on sample data. The interval provides an estimate of where the true population parameter is likely to fall, given a certain level of confidence.

A confidence interval is not a precise value, but rather a range, and it is associated with a confidence level (e.g., 95%, 99%) that expresses how confident we are that the interval contains the true parameter.

The margin of error (MOE) is a measure of the uncertainty or variability in an estimate.
```python
#Practical
#1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results?
import numpy as np
import scipy.stats as stats
def z_test(sample, pop_mean, pop_std, alpha=0.05, alternative="two-sided"):
    """
    Perform a Z-test for comparing a sample mean to a known population mean.

    Parameters:
    sample (list or array): Sample data
    pop_mean (float): Population mean
    pop_std (float): Population standard deviation (known)
    alpha (float): Significance level (default is 0.05)
    alternative (str): "two-sided", "greater", or "less"

    Returns:
    Z-score, p-value, and interpretation of the results
    """

    # Sample statistics
    sample_mean = np.mean(sample)
    sample_size = len(sample)

    # Compute Z-score
    z_score = (sample_mean - pop_mean) / (pop_std / np.sqrt(sample_size))

    # Compute p-value based on test type
    if alternative == "two-sided":
        p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))
    elif alternative == "greater":
        p_value = 1 - stats.norm.cdf(z_score)
    elif alternative == "less":
        p_value = stats.norm.cdf(z_score)
    else:
        raise ValueError("Invalid alternative hypothesis. Choose 'two-sided', 'greater', or 'less'.")

    # Interpretation
    result = "Reject H₀" if p_value < alpha else "Fail to reject H₀"

    return {
        "Z-score": round(z_score, 4),
        "p-value": round(p_value, 4),
        "Conclusion": result
    }

# Example Usage
sample_data = [72, 74, 71, 69, 73, 75, 70, 72, 74, 71]  # Sample data
population_mean = 70  # Known population mean
population_std = 3  # Known population standard deviation

# Perform Z-test (two-tailed test)
result = z_test(sample_data, population_mean, population_std, alpha=0.05, alternative="two-sided")

# Display results
print("Z-Test Results:")
print(result)
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing?
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def visualize_type1_type2_errors(pop_mean, pop_std, sample_size, effect_size, alpha=0.05):
    """
    Plots Type 1 (α) and Type 2 (β) errors for a hypothesis test.

    Parameters:
    - pop_mean: float, true population mean under H0
    - pop_std: float, population standard deviation
    - sample_size: int, number of observations
    - effect_size: float, difference between H0 mean and true mean under H1
    - alpha: float, significance level (default 0.05)
    """

    # Compute standard error
    standard_error = pop_std / np.sqrt(sample_size)

    # Define null and alternative means
    h0_mean = pop_mean
    h1_mean = pop_mean + effect_size  # Mean under H1

    # Compute critical value for given alpha (two-tailed test)
    z_critical = stats.norm.ppf(1 - alpha / 2)
    critical_value = h0_mean + z_critical * standard_error

    # Define x values for distribution curves
    x = np.linspace(h0_mean - 4 * standard_error, h1_mean + 4 * standard_error, 1000)

    # Compute probability density functions
    h0_pdf = stats.norm.pdf(x, h0_mean, standard_error)  # H0 distribution
    h1_pdf = stats.norm.pdf(x, h1_mean, standard_error)  # H1 distribution

    plt.figure(figsize=(10, 5))

    # Plot distributions
    plt.plot(x, h0_pdf, label="H₀ Distribution (Null Hypothesis)", color="blue")
    plt.plot(x, h1_pdf, label="H₁ Distribution (Alternative Hypothesis)", color="green")

    # Shade Type 1 Error (α) - Rejection region under H0
    plt.fill_between(x, h0_pdf, where=(x >= critical_value), color="red", alpha=0.3, label="Type 1 Error (α)")

    # Shade Type 2 Error (β) - Acceptance region under H1
    beta = stats.norm.cdf(critical_value, h1_mean, standard_error)
    plt.fill_between(x, h1_pdf, where=(x < critical_value), color="orange", alpha=0.3, label=f"Type 2 Error (β ≈ {beta:.3f})")

    # Plot critical value
    plt.axvline(critical_value, color="red", linestyle="--", label=f"Critical Value ({critical_value:.2f})")

    # Labels and legend
    plt.title("Type 1 (α) and Type 2 (β) Errors in Hypothesis Testing")
    plt.xlabel("Sample Mean")
    plt.ylabel("Probability Density")
    plt.legend()
    plt.show()

# Example Usage
visualize_type1_type2_errors(pop_mean=100, pop_std=15, sample_size=30, effect_size=5, alpha=0.05)
```

```python
#6.Write a Python program to perform an independent T-test and interpret the results?


import numpy as np
import scipy.stats as stats

def independent_t_test(sample1, sample2, alpha=0.05):
    """
    Performs an independent (two-sample) T-test.

    Parameters:
    - sample1: list or numpy array, data for group 1
    - sample2: list or numpy array, data for group 2
    - alpha: float, significance level (default 0.05)

    Returns:
    - T-statistic, p-value, and conclusion of the hypothesis test
    """

    # Check for equal variances using Levene’s test
    levene_stat, levene_p = stats.levene(sample1, sample2)
    equal_var = levene_p > alpha  # Assume equal variance if p > alpha

    # Perform independent T-test
    t_stat, p_value = stats.ttest_ind(sample1, sample2, equal_var=equal_var)

    # Conclusion
    conclusion = "Reject H₀" if p_value < alpha else "Fail to reject H₀"

    return {
        "T-statistic": round(t_stat, 4),
        "p-value": round(p_value, 4),
        "Equal Variance Assumed": equal_var,
        "Levene’s Test p-value": round(levene_p, 4),
        "Conclusion": conclusion
    }

# Example Usage
np.random.seed(42)  # For reproducibility
group1 = np.random.normal(loc=100, scale=15, size=30)  # Simulated group 1 data
group2 = np.random.normal(loc=105, scale=15, size=30)  # Simulated group 2 data

# Perform Independent T-Test
result = independent_t_test(group1, group2, alpha=0.05)

# Print the result
print("Independent T-Test Results:")
print(result)
```

Bayes' Theorem in Statistics

Bayes' Theorem is a fundamental concept in probability theory and statistics that describes how to update the probability of a hypothesis (or event) based on new evidence. It provides a way to revise our beliefs about the likelihood of an event, given prior knowledge and observed data.

In mathematical terms, Bayes' Theorem is:

P(E∣H) is the likelihood, the probability of observing the evidence 𝐸 E given that the hypothesis 𝐻 H is true.
𝑃 ( 𝐻 ) P(H) is the prior probability, the initial probability of the hypothesis 𝐻 H before seeing the evidence 𝐸 E.

∑ 𝐻 𝑃 ( 𝐸 ∣ 𝐻 ) × 𝑃 ( 𝐻 ) P(E)= H ∑​P(E∣H)×P(H)
How Bayes' Theorem Works

Bayes' Theorem allows you to update your beliefs about a hypothesis or event as new evidence or data becomes available. Here’s how it works in the context of statistics:

Prior probability ( 𝑃 ( 𝐻 ) P(H)): Before observing new data, we have a prior belief or assumption about the probability of a hypothesis being true.

Likelihood ( 𝑃 ( 𝐸 ∣ 𝐻 ) P(E∣H)): After observing new data, we evaluate how likely it is to observe this evidence given the hypothesis.

Posterior probability ( 𝑃 ( 𝐻 ∣ 𝐸 ) P(H∣E)): Using Bayes' Theorem, we combine our prior belief with the likelihood of the evidence to calculate the updated probability (posterior) of the hypothesis.

Significance of Bayes' Theorem

Bayes' Theorem is extremely significant for several reasons:

Updating Beliefs with New Data
Bayesian updating allows you to continuously refine your knowledge as new data comes in. It is especially useful in situations where we start with an initial assumption (prior probability) and want to adjust it as more evidence becomes available.
Real-World Applications

Spam Filtering: Bayes' Theorem is the foundation of Bayesian spam filters, which classify emails as either spam or non-spam based on prior probabilities of spam emails and the likelihood of certain words appearing in them.

Machine Learning: Bayesian methods are widely used in machine learning and data science for updating predictive models based on new data. Bayesian networks and Markov chain Monte Carlo (MCMC) methods rely on Bayes' Theorem for probabilistic inference.

Handling Uncertainty
In many real-world problems, there is inherent uncertainty. Bayes' Theorem helps model and quantify this uncertainty, allowing for more flexible decision-making based on partial or incomplete information.
Incorporating Prior Knowledge
The prior in Bayes' Theorem allows you to incorporate previous knowledge or expert opinion into the analysis. This is especially important when data is scarce or when the sample size is small.

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance?


import numpy as np
import scipy.stats as stats

def confidence_interval(sample, confidence=0.95):
    """
    Calculates the confidence interval for the mean of a sample.

    Parameters:
    - sample: list or numpy array, the sample data
    - confidence: float, confidence level (default 0.95)

    Returns:
    - (lower_bound, upper_bound): Tuple containing the confidence interval bounds
    """
    sample_mean = np.mean(sample)
    sample_std = np.std(sample, ddof=1)  # Sample standard deviation (unbiased)
    n = len(sample)

    # Select t-distribution or Z-distribution based on sample size
    if n >= 30:
        z_or_t = stats.norm.ppf(1 - (1 - confidence) / 2)  # Z-score for large samples
    else:
        z_or_t = stats.t.ppf(1 - (1 - confidence) / 2, df=n-1)  # t-score for small samples

    margin_of_error = z_or_t * (sample_std / np.sqrt(n))

    lower_bound = round(sample_mean - margin_of_error, 4)
    upper_bound = round(sample_mean + margin_of_error, 4)

    return lower_bound, upper_bound

# Example Usage
np.random.seed(42)  # For reproducibility
sample_data = np.random.normal(loc=100, scale=15, size=25)  # Simulated sample

ci = confidence_interval(sample_data, confidence=0.95)
print(f"95% Confidence Interval: {ci}")
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process?


def bayes_theorem(prior, sensitivity, false_positive_rate):
    """
    Implements Bayes' Theorem for a binary classification problem.

    Parameters:
    - prior: float, P(H) - Prior probability of the hypothesis (e.g., having COVID-19).
    - sensitivity: float, P(E | H) - Sensitivity or True Positive Rate.
    - false_positive_rate: float, P(E | ~H) - False Positive Rate.

    Returns:
    - posterior: float, P(H | E) - Updated probability after observing evidence.
    """
    # P(E) = P(E | H) * P(H) + P(E | ~H) * P(~H)
    marginal_likelihood = (sensitivity * prior) + (false_positive_rate * (1 - prior))

    # P(H | E) = (P(E | H) * P(H)) / P(E)
    posterior = (sensitivity * prior) / marginal_likelihood

    return round(posterior, 4)

# Example: COVID-19 Test
prior_covid = 0.01  # 1% of people have COVID
sensitivity_covid = 0.95  # True positive rate
false_positive_covid = 0.05  # False positive rate

posterior_covid = bayes_theorem(prior_covid, sensitivity_covid, false_positive_covid)
print(f"Probability of having COVID-19 after testing positive: {posterior_covid * 100:.2f}%")
```

#19.Write a Python function to check the assumptions (normality, independence, and equal variance) for ANOVA?


import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def check_assumptions(*groups):
    """
    Check the assumptions for performing ANOVA (normality, independence, equal variance).

    Parameters:
    - groups: The groups to be checked (each group is a list or numpy array)

    Returns:
    - assumption_results: A dictionary with the results of the assumption checks
    """

    assumption_results = {}

    # 1. Normality Assumption - Shapiro-Wilk Test
    normality_results = {}
    for i, group in enumerate(groups):
        stat, p_value = stats.shapiro(group)  # Shapiro-Wilk test for normality
        normality_results[f"Group {i+1}"] = (stat, p_value)

    # Check if p-value is greater than 0.05 (fail to reject H0: the group is normal)
    assumption_results["Normality"] = {k: ("Normal" if p > 0.05 else "Not Normal") for k, (stat, p) in normality_results.items()}

    # 2. Equal Variance Assumption - Levene's Test
    stat, p_value = stats.levene(*groups)  # Levene's test for equal variances
    assumption_results["Equal Variance (Levene's Test)"] = ("Equal Variance" if p_value > 0.05 else "Not Equal Variance")

    # 3. Independence Assumption
    # Independence cannot be tested statistically, but we can assume independence if data is randomly sampled
    assumption_results["Independence"] = "Assumed Random Sampling"

    return assumption_results
